* 0, or 'index' ⁠:⁠ Drop rows which contain missing values.

```python
# Load data from a .csv file
!wget -O grades.csv https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/grades.csv
df_students = pd.read_csv('grades.csv', delimiter=',', header= 'infer')

# Remove any rows with missing values
df_students = df_students.dropna(axis=0, how='any')

# calculate who passed, assuming 60 is the passing grade or the neede grade to pass
passes = pd.Series(df_students.Grade >= 60) 
# passes = pd.Series(df_students['Grade'] >= 60)

# save who passed to the Pandasd dataframe
df_students = pd.concat([df_students, passes.rename('Passed')], axis=1)

# Print the result out into this notebook
df_students 

```

```python
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(12.8, 7.2))
plt.bar(x=df_students.Name, height=df_students.Grade, color='orange')
plt.title('Student Grades')
plt.xlabel('Student Name')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.ylabel('Grade')
plt.show()
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].bar(x=df_students.Name, height=df_students.Grade, color='orange')
ax[0].set_title('Student Grades')
# ax[0].set_xlabel('Student Name')
ax[0].set_xticklabels(df_students.Name, rotation=90)  # Rotate x-axis labels for better readability
passes_count = df_students.Passed.value_counts()
ax[1].pie(passes_count, labels=passes_count)
ax[1].set_title('Passing Grades')
fig.show()
```

```python
# Central Tendencies, Mean, Median, and Mode
grades = df_students.Grade
min_grade = grades.min()
max_grade = grades.max()
mean_grade = grades.mean()
median_grade = grades.median()
mode_grade = grades.mode()[0]

print(f'Minimum Grade: {min_grade}, Maximum Grade: {max_grade}, Mean Grade: {mean_grade:.2f}, Median Grade: {median_grade}, Mode Grade: {mode_grade}')  

fig = plt.figure(figsize=(10, 4))
# Plot the boxplot
plt.hist(grades) #, bins=10, color='lightGreen', edgecolor='green')

# Add lines for the mean, median, and mode
plt.axvline(min_grade, color='gray', linestyle='dashed', linewidth=1, label=f'Min: {min_grade}')
plt.axvline(max_grade, color='gray', linestyle='dashed', linewidth=1, label=f'Max: {max_grade}')
plt.axvline(mean_grade, color='yellow', linestyle='dashed', linewidth=2, label=f'Mean: {mean_grade:.2f}')
plt.axvline(median_grade, color='red', linestyle='dashed', linewidth=2, label=f'Median: {median_grade}')
plt.axvline(mode_grade, color='black', linestyle='solid', linewidth=1, label=f'Mode: {mode_grade}')

# Add a title to the Figure
plt.title('Distribution of Student Grades')

plt.xlabel('Grade')
plt.ylabel('Frequency')

# Show the figure
plt.show()
```

## boxplot 
A boxplot shows the distribution of a numeric column in a compact way. 
In pandas, it typically displays 
* the minimum, 
* first quartile, 
* median, 
* third quartile, 
* maximum, 
* and possible outliers.
  
A boxplot helps quickly see:
* where most values lie,
* how spread out the data is,
* whether the data is skewed,
* whether there are outliers

```python
# Creating a boxplot with Matplotlib
grades = df_students.Grade
fig = plt.figure(figsize=(10, 4))
plt.boxplot(grades)
plt.title('Boxplot of Student Grades')
plt.show()
```
